```r
library(tidyverse)
library(tidytext)
library(glue)
library(stringr)
```

```r
data <- read.csv('india-news-headlines.csv')
```

```r
library(lubridate)
```

```r
data$publish_date <- ymd(data$publish_date)
```

```r
data
```

```r
sum(table(data$headline_category) > 10)
```

```r
head_count <- count(group_by(data, data$headline_category))
```

```r
head_count <- head_count[head_count$n >= 10,]
```

```r
head_count[order(head_count$n),]
```

```r
data2 <- data[which(head_count["data$headline_category"] != 'tv.news.hindi'),]
```

```r
data2 <- filter(data, headline_category != 'tv.news.hindi')
```

```r
table(year(data2$publish_date))
```

```r
library(dplyr)
```

```r
library(data.table)
```

```r
temp <- as.data.table(data2)
temp$headline_text <- as.character(temp$headline_text)
```

```r
table(year(temp$publish_date))
```

```r
str(temp)
```

```r
temp$set_obj <- get_sentences(temp$headline_text)
```

```r
temp$sentiment <- sentiment_by(temp$set_obj)$ave_sentiment
```

```r
head(temp)
```

```r
year_sentiment <- temp[,mean(sentiment), by = .(year(publish_date))]
```

```r
year_sentiment
```

```r
month_sentiment <- temp[,mean(sentiment), by = .(month(publish_date))]
```

```r
month_sentiment
```

```r
cat_sentiment <- temp[,mean(sentiment), by = .(headline_category)]
```

```r
cat_sentiment <- cat_sentiment[order(cat_sentiment$V1),]
```

```r
filter(temp, headline_category == 'india')
```
